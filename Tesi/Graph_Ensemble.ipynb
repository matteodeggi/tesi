```python
import numpy as np
```

```python
metrics_things = [[75.8, ['DecisionTreeClassifier', 'SVC', 'KNeighborsClassifier', 'LogisticRegression', 'GradientBoostingClassifier']],
                  [76.5, [['RandomForest'], ['GradientBoostingClassifier']]],
                  [75.51, ['KNeighborsClassifier', 'RandomForestClassifier', 'DecisionTreeClassifier']],
                  [75.67, ['DecisionTreeClassifier', 'KNeighborsClassifier', 'GradientBoostingClassifier']]]
```

```python
metrics_swissex = [[79.8, ['RandomForestClassifier', 'DecisionTreeClassifier', 'GaussianNB']],
                  [76.9, ['RandomForest', 'GaussianNB', 'GradBoost']],
                  [77.885, ['GradientBoostingClassifier', 'LogisticRegression']],
                  [79.808, ['GaussianNB', 'GradientBoostingClassifier']]]
```

```python
metrics_urban = [[89.4, ['RandomForestClassifier', 'LogisticRegression']],
                [88.1, [['RandomForest'], ['KNeighborsClassifier']]],
                [88.75, ['RandomForestClassifier', 'DecisionTreeClassifier', 'GaussianNB']],
                [88.75, ['GradientBoostingClassifier', 'GaussianNB', 'RandomForestClassifier']]]
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=(20,10))
index = np.arange(4)
bar_width = 0.25

labels = ['Voting', 'Stacking', 'SoF', 'TopK']
opacity = .6
error_config = {'ecolor': '0.3'}
    
urban_ax = ax.bar(index - bar_width, [y[0] for y in metrics_urban], bar_width, alpha=opacity, color='g', error_kw=error_config, label='UrbanObservatory')
swissex_ax = ax.bar(index, [y[0] for y in metrics_swissex], bar_width, alpha=opacity, color='c', error_kw=error_config, label='Swissex')
things_ax = ax.bar(index + bar_width, [y[0] for y in metrics_things], bar_width, alpha=opacity, color='#BA3B46', error_kw=error_config, label='Thingspeak')
    
def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{:.2f}%'.format(round(height, 4)),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom', rotation = 90)
autolabel(things_ax)
autolabel(swissex_ax)
autolabel(urban_ax)
    
ax.set_title('Accuracy a confronto su 3 dataset per algoritmi ensemble che operano su dati numerici')
ax.set_ylim(0, 100)
ax.set_xlabel('Classifiers')
ax.set_ylabel('Accuracy')
ax.set_xticks(index)
ax.set_xticklabels((labels))
ax.legend()
plt.savefig('SingleDomain_Ensemble.png', bbox_inches='tight',pad_inches = 0, dpi = 200)
plt.show()
```

```python
import numpy as np
import pandas as pd
```

```python
metrics_things = [[71.27, 75.09],
 [71.27, 74.77],
 [70.80, 67.47],
 [70.33, 68.74]]
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=(20,10))
index = np.arange(4)
bar_width = 0.35

labels = ['Damerau-Levenshtein', 'Levenshtein', 'Jaccard', 'Jaro-Winkler']
opacity = .7
error_config = {'ecolor': '0.3'}
    
things_ax = ax.bar(index, [y[0] for y in metrics_things], bar_width, alpha=opacity, color='#BA3B46', error_kw=error_config, label='Thingspeak')
    
def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{:.2f}%'.format(round(height, 4)),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
autolabel(things_ax)
    
ax.set_title('Accuracy a confronto per diverse edit-distances')
ax.set_ylim(0, 100)
ax.set_xlabel('Classifiers')
ax.set_ylabel('Accuracy')
ax.set_xticks(index)
ax.set_xticklabels((labels))
ax.legend()
plt.savefig('NLP_Algorithms.png', bbox_inches='tight',pad_inches = 0, dpi = 200)
plt.show()
```

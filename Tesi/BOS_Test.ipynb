```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=(20,10))
index = np.arange(8)
bar_width = 0.3

labels = ['DecisionTree', 'SVM', 'KNN', 'Logistic', 'Ridge', 'Naive Bayes', 'Random Forest', 'Gradient Boosting']
opacity = .7
error_config = {'ecolor': '0.3'}
    
urban_ax = ax.bar(index - bar_width, [y[0] for y in metrics_urban], bar_width, alpha=opacity, color='g', error_kw=error_config, label='UrbanObservatory')
swissex_ax = ax.bar(index, [y[0] for y in metrics_swissex], bar_width, alpha=opacity, color='c', error_kw=error_config, label='Swissex')
things_ax = ax.bar(index + bar_width, [y[0] for y in metrics_things], bar_width, alpha=opacity, color='#BA3B46', error_kw=error_config, label='Thingspeak')
    
def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{:.2f}%'.format(round(height, 4)),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
autolabel(things_ax)
autolabel(swissex_ax)
autolabel(urban_ax)
    
ax.set_title('Accuracy a confronto su 3 dataset')
ax.set_ylim(0, 100)
ax.set_xlabel('Classifiers')
ax.set_ylabel('Accuracy')
ax.set_xticks(index)
ax.set_xticklabels((labels))
ax.legend()
plt.savefig('BOS_Algorithms.png', bbox_inches='tight',pad_inches = 0, dpi = 200)
plt.show()
```
